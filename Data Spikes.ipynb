```python
import pandas as pd
import matplotlib.pyplot as plt

def find_large_spikes(data, column, threshold):
    """
    Identifies large spikes in the specified column of a dataset, includes the day before and after each spike,
    removes non-unique rows, and plots the results.

    Parameters:
    - data (pd.DataFrame): The dataset containing the data.
    - column (str): The column name in which to look for spikes (e.g., 'GWAP' or 'LWAP').
    - threshold (float): The minimum change between consecutive values to consider as a spike.

    Returns:
    - pd.DataFrame: A DataFrame with unique RUN_TIME, the spike values, the day before, and the day after each spike.
    """
    # Calculate the absolute change between consecutive values in the specified column
    data['Change'] = data[column].diff().abs()
    
    # Find rows where the change exceeds the threshold
    spike_indices = data.index[data['Change'] > threshold]
    
    # Initialize an empty list to store rows with the day before, spike day, and day after
    spike_rows = []
    for idx in spike_indices:
        # Get the day before, spike day, and day after if within bounds
        if idx > 0 and idx < len(data) - 1:
            spike_rows.append(data.iloc[idx - 1: idx + 2])
    
    # Concatenate the spike_rows list into a single DataFrame and drop duplicates based on RUN_TIME
    spike_data = pd.concat(spike_rows).drop_duplicates(subset='RUN_TIME').reset_index(drop=True)
    
    # Plot the time series and highlight detected spikes
    plt.figure(figsize=(10, 6))
    plt.plot(data['RUN_TIME'], data[column], label=column, color='blue')
    plt.scatter(data.iloc[spike_indices]['RUN_TIME'], data.iloc[spike_indices][column], color='orange', label='Detected Spikes', zorder=5)
    plt.xlabel('RUN_TIME')
    plt.ylabel(column)
    plt.title(f'Spike Detection in {column} with Threshold {threshold}')
    plt.legend()
    plt.show()
    
    # Return unique rows of RUN_TIME and the spike values, along with the spike amount and adjacent days
    return spike_data[['RUN_TIME', column, 'Change']]

# Example usage:
# df = pd.DataFrame({'RUN_TIME': [...], 'GWAP': [...]})
# threshold_value = 10  # Define the spike threshold based on your data
# large_spikes = find_large_spikes(df, 'GWAP', threshold_value)
# print(large_spikes)
```

```python
import pandas as pd

def top_product_per_day(data):
    """
    This function performs the following steps:
    1. Adds a new column 'PRODUCT' which is the product of 'LMP' and 'SCHED_MW' for each row.
    2. Sorts the 'PRODUCT' values per day in descending order.
    3. Displays the top 20 'PRODUCT' values per day.

    Parameters:
    - data (pd.DataFrame): The input DataFrame with 'TIME_INTERVAL', 'LMP', and 'SCHED_MW' columns.

    Returns:
    - pd.DataFrame: A DataFrame with the top 20 'PRODUCT' values per day.
    """
    # Ensure TIME_INTERVAL is in datetime format if it's not already
    data = data.copy()  # Create a full copy to avoid the SettingWithCopyWarning
    data['TIME_INTERVAL'] = pd.to_datetime(data['TIME_INTERVAL'])
    
    # Step 1: Create a new column 'PRODUCT' as the product of 'LMP' and 'SCHED_MW'
    data['PRODUCT'] = data['LMP'] * data['SCHED_MW']
    
    # Step 2: Sort the data by 'TIME_INTERVAL' (day) and 'PRODUCT' in descending order
    data['Date'] = data['TIME_INTERVAL'].dt.date  # Extract the date part for grouping
    
    # Sort by Date and PRODUCT
    sorted_data = data.sort_values(by=['Date', 'PRODUCT'], ascending=[True, False])
    
    # Step 3: Group by date and take the top 20 'PRODUCT' values per day
    top_20_per_day = sorted_data.groupby('Date').head(20).reset_index(drop=True)
    
    # Display the results
    print("Top 20 'PRODUCT' values per day:")
    
    return top_20_per_day[['Date', 'TIME_INTERVAL', 'RESOURCE_NAME', 'LMP', 'SCHED_MW', 'PRODUCT']]

# Example usage:
# Assuming you have a DataFrame 'cluz_data' with the necessary columns
# top_20_products = top_product_per_day(cluz_data)
```

```python
import pandas as pd

def unique_resource_names_per_day(data):
    """
    This function groups data by each day and extracts the unique RESOURCE_NAME values for each day.

    Parameters:
    - data (pd.DataFrame): The input DataFrame with 'TIME_INTERVAL' and 'RESOURCE_NAME' columns.

    Returns:
    - pd.DataFrame: A DataFrame with each day and the corresponding list of unique RESOURCE_NAME values.
    """
    # Ensure TIME_INTERVAL is in datetime format
    data = data.copy()  # Make a copy to avoid modifying the original DataFrame
    data['TIME_INTERVAL'] = pd.to_datetime(data['TIME_INTERVAL'])
    
    # Extract date part only for grouping
    data['Date'] = data['TIME_INTERVAL'].dt.date
    
    # Group by Date and get unique RESOURCE_NAME values for each day
    unique_resources = data.groupby('Date')['RESOURCE_NAME'].unique().reset_index()
    
    # Rename the column for clarity
    unique_resources.rename(columns={'RESOURCE_NAME': 'Unique_Resource_Names'}, inplace=True)
    
    # Adjust display options to show the full content of each cell
    pd.set_option('display.max_colwidth', None)
    
    # Display the results
    print("Unique RESOURCE_NAME values per day:")
    print(unique_resources)
    
    # Reset display options to default after printing
    pd.reset_option('display.max_colwidth')
    
    return unique_resources

# Example usage:
# Assuming you have a DataFrame 'cluz_data' with the necessary columns
# unique_resources_per_day = unique_resource_names_per_day(cluz_data)
```
