```python
# Get unique values in COMMODITY_TYPE column
unique_commodities = df_reserve['COMMODITY_TYPE'].unique()

# Create an empty dictionary to store the new DataFrames
commodity_dfs = {}

# Loop through each unique commodity type and create a DataFrame for each
for commodity in unique_commodities:
    # Create a new DataFrame for each commodity type
    commodity_dfs[commodity] = df_reserve[df_reserve['COMMODITY_TYPE'] == commodity]

    # Optional: Print the first few rows of each new DataFrame
    print(f"DataFrame for {commodity}:")
    display(commodity_dfs[commodity])
```

```python
# Loop through each DataFrame in the commodity_dfs dictionary
for commodity, df_commodity in commodity_dfs.items():
    # Filter out rows where RESOURCE_TYPE is 'PS'
    commodity_dfs[commodity] = df_commodity[df_commodity['RESOURCE_TYPE'] != 'PS']
    print(df_commodity['RESOURCE_TYPE'].unique())
    print(df_commodity['REGION_NAME'].unique())
```

```python
import pandas as pd

# Assuming commodity_dfs contains the data frames by COMMODITY_TYPE
# Loop through each DataFrame in commodity_dfs and process them
for commodity, df_commodity in commodity_dfs.items():
    # Step 1: Create separate DataFrames for each REGION_NAME within the commodity DataFrame
    regions = df_commodity['REGION_NAME'].unique()

    for region in regions:
        # Step 1: Filter the data for the current region using loc
        region_df = df_commodity.loc[df_commodity['REGION_NAME'] == region].copy()  # Use .copy() to avoid SettingWithCopyWarning

        # Step 2: Aggregate 5-minute interval data into daily data using a weighted average
        # Convert RUN_TIME to datetime, if not already
        region_df.loc[:, 'RUN_TIME'] = pd.to_datetime(region_df['RUN_TIME'])
        
        # Set RUN_TIME as the index
        region_df.set_index('RUN_TIME', inplace=True)
        
        # Calculate the weighted average (sumproduct of PRICE and SCHED_MW divided by total SCHED_MW)
        def weighted_avg(x):
            total_sched_mw = x['SCHED_MW'].sum()
            if total_sched_mw == 0:
                return float('nan')  # or return 0 if preferred
            return (x['PRICE'] * x['SCHED_MW']).sum() / total_sched_mw

        daily_aggregated = region_df.resample('D').apply(weighted_avg)

        # Step 3: Reset the index so that RUN_TIME becomes a column again
        daily_aggregated = daily_aggregated.reset_index()

        # Step 4: Rename the columns dynamically based on the commodity
        dynamic_column_name = f'RESERVE_GWAP_{commodity}'
        daily_aggregated.columns = ['RUN_TIME', dynamic_column_name]

        # Step 5: Print the head of each new DataFrame
        print(f"Commodity: {commodity}, Region: {region}")
        print(daily_aggregated.head())

        # Step 6: Save the DataFrame to a CSV file with the format Reserve_Daily_{commodity}.csv
        output_filename = f'{region}_Reserve_Daily_{commodity}_.csv'
        daily_aggregated.to_csv(output_filename, index=False)

        print(f"Saved CSV for {commodity} in region {region} as {output_filename}")
```

```python
df_lwap = pd.read_csv('daily_LWAP_trial.csv')
df_gwap = pd.read_csv('daily_GWAP_trial.csv')
print(df_lwap.head())
print(df_gwap.head())
```

```python
# Rearrange the columns for df_lwap
df_lwap = df_lwap[['RUN_TIME', 'LWAP', 'REGION_NAME']]

# Rearrange the columns for df_gwap
df_gwap = df_gwap[['RUN_TIME', 'GWAP', 'REGION_NAME']]

# Print the updated DataFrames
print(df_lwap.head())
print(df_gwap.head())
```

```python
import pandas as pd

# Create an empty dictionary to store the new DataFrames
region_dfs = {}

# Get unique region names from df_lwap and df_gwap
unique_regions_lwap = df_lwap['REGION_NAME'].unique()
unique_regions_gwap = df_gwap['REGION_NAME'].unique()

# Define the datetime format if you know it
datetime_format = '%m/%d/%y'

# For df_lwap: Create DataFrames for each unique region and store them in the dictionary
for region in unique_regions_lwap:
    # Filter the DataFrame for the current region
    df_region_lwap = df_lwap[df_lwap['REGION_NAME'] == region].copy()
    
    # Convert RUN_TIME to datetime format with specified format
    df_region_lwap['RUN_TIME'] = pd.to_datetime(df_region_lwap['RUN_TIME'], format=datetime_format)
    
    # Remove rows from 2021
    df_region_lwap = df_region_lwap[df_region_lwap['RUN_TIME'].dt.year != 2021]
    
    # Drop the REGION_NAME column
    df_region_lwap.drop(columns=['REGION_NAME'], inplace=True)
    
    # Store in dictionary with the key as 'LWAP_{region}'
    region_dfs[f'LWAP_{region}'] = df_region_lwap

# For df_gwap: Create DataFrames for each unique region and store them in the dictionary
for region in unique_regions_gwap:
    # Filter the DataFrame for the current region
    df_region_gwap = df_gwap[df_gwap['REGION_NAME'] == region].copy()
    
    # Convert RUN_TIME to datetime format with specified format
    df_region_gwap['RUN_TIME'] = pd.to_datetime(df_region_gwap['RUN_TIME'], format=datetime_format)
    
    # Remove rows from 2021
    df_region_gwap = df_region_gwap[df_region_gwap['RUN_TIME'].dt.year != 2021]
    
    # Drop the REGION_NAME column
    df_region_gwap.drop(columns=['REGION_NAME'], inplace=True)
    
    # Store in dictionary with the key as 'GWAP_{region}'
    region_dfs[f'GWAP_{region}'] = df_region_gwap

# Save each DataFrame to a CSV file with the naming convention Daily_{key}.csv
for key, df in region_dfs.items():
    # Construct the filename
    output_filename = f'Daily_{key}.csv'
    
    # Save the DataFrame as a CSV
    df.to_csv(output_filename, index=False)
    
    # Print a message to confirm saving
    print(f"Saved {output_filename}")
```
